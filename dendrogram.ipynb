```python
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt
```

```python
data=[ [-0.1000 ,  -0.3000],
   [-1.9000 ,   2.6000],
   [ 0.9000  ,  2.6000],
   [-1.0000   ,-2.2000],
   [ 1.6000    ,0.1000]]
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt
Z = linkage(data, 'single')
fig = plt.figure()
dn = dendrogram(Z)
plt.title('single method')
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt
X = [[i] for i in [2,6.5,2.5,10,11,4,3,11.5,7,12]]
Z = linkage(X, 'complete')
fig = plt.figure()
dn = dendrogram(Z)
plt.show()
```
